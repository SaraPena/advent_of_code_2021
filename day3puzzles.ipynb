```python
for x in example_df.columns[0:7]:
    print(x)
    if x == 0:
        ex_zeros = len(example_df) - example_df.sum()
        ex_ones = example_df.sum()
        ex_zip_list = [x for x in zip(ex_zeros,ex_ones)]
        if ex_zip_list[x][0]>ex_zip_list[x][1]:
            new_df = example_df[example_df[x]==0]
        else:
            new_df = example_df[example_df[x]==1]
            
    else:
        ex_zeros = len(new_df) - new_df.sum()
        ex_ones = new_df.sum()
        ex_zip_list = [x for x in zip(ex_zeros,ex_ones)]
        if ex_zip_list[x][0]>ex_zip_list[x][1]:
            new_df = new_df[new_df[x]==0]
        else:
            new_df = new_df[new_df[x]==1]
        
len(new_df),new_df
```

```python
# calculate oxygen generator rating
for x in final_df.columns:
    print(x)
    if x == 0:
        zeros = len(final_df) - final_df.sum()
        ones = final_df.sum()
        zip_list = [x for x in zip(zeros, ones)]
        if zip_list[x][0]>zip_list[x][1]:
            print('most common: 0')
            new_df = final_df[final_df[x]==0]
        else:
            if zip_list[x][0]==zip_list[x][1]:
                print('both equal - keep 1')
            else:
                print('most common: 1')
            new_df = final_df[final_df[x]==1]
            
    else:
        zeros = len(new_df) - new_df.sum()
        ones = new_df.sum()
        zip_list = [x for x in zip(zeros,ones)]
        if zip_list[x][0]>zip_list[x][1]:
            print('most common: 0')
            new_df = new_df[new_df[x]==0]
        else:
            if zip_list[x][0]==zip_list[x][1]:
                print('both equal - keep 1')
            else:
                print('most common: 1')
            new_df = new_df[new_df[x]==1]
        
oxygen_generator_rating = int(''.join([str(x) for x in new_df.iloc[0]]),base=2)
oxygen_generator_rating
```

```python
# calculate CO2 scrubber rating rating
for x in final_df.columns[0:9]:
    print(x)
    if x == 0:
        zeros = len(final_df) - final_df.sum()
        ones = final_df.sum()
        zip_list = [x for x in zip(zeros, ones)]
        print(f'zeros:{zip_list[x][0]}',f'ones:{zip_list[x][1]}')

        if zip_list[x][0]<zip_list[x][1]:
            print('least common: 0')
            new_df = final_df[final_df[x]==0]
            print(f'new_df_length:{len(new_df)}')

        else:
            if zip_list[x][0]==zip_list[x][1]:
                print('both equal - keep 0')
                new_df = final_df[final_df[x]==0]

            else:
                print('least common: 1')
                new_df = final_df[final_df[x]==1]
            print(f'new_df_length:{len(new_df)}')
            
    else:
        zeros = len(new_df) - new_df.sum()
        ones = new_df.sum()
        zip_list = [x for x in zip(zeros,ones)]
        print(f'zeros:{zip_list[x][0]}',f'ones:{zip_list[x][1]}')
        if zip_list[x][0]<zip_list[x][1]:
            print('least common: 0')
            new_df = new_df[new_df[x]==0]
            print(f'new_df_length:{len(new_df)}')

        else:
            if zip_list[x][0]==zip_list[x][1]:
                print('both equal - keep 0')
                new_df = new_df[new_df[x]==0]

            else:
                print('least common: 1')
                new_df = new_df[new_df[x]==1]
            print(f'new_df_length:{len(new_df)}')

CO2_scrubber_rating = int(''.join([str(x) for x in new_df.iloc[0]]),base=2)
CO2_scrubber_rating
```
